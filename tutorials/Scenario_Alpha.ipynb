## Uncontrolled Response

```python
env = pystorms.scenarios.alpha()
done = False
while not done:
    done = env.step()
```

```python
flows = pd.DataFrame(index=env.data_log['simulation_time'], data=env.data_log['flow'])
```

```python
plt.figure(figsize=(10, 5), dpi=100)
ax = plt.gca()
flows.loc[:datetime(2007, 1, 1, 1, 30, 0)].plot(ax=ax)
plt.title("Flows from Weirs")
plt.ylabel("Flow $m^3s^{-1}$")
plt.xlabel("Simulation Time")
```
